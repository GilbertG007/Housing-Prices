```python
# Extract lettable space and service charge data correctly
space_charge_data = {
    'Year': ['2022', '2023', '2024'],
    'Lettable Space (sqft)': [95400, 101196, 101196],  # From your Excel data
    'Service Charge per sqft/year': [231.9, 260.75, 206.16]  # From your Excel data
}
space_df = pd.DataFrame(space_charge_data).set_index('Year')

# Create the comparison plot
fig, ax1 = plt.subplots(figsize=(10, 6))

color = 'tab:blue'
ax1.set_xlabel('Year', fontsize=12)
ax1.set_ylabel('Lettable Space (sqft)', color=color, fontsize=12)
ax1.plot(space_df.index, space_df['Lettable Space (sqft)'],
         color=color, marker='o', linewidth=2)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Service Charge (per sqft/year)', color=color, fontsize=12)
ax2.plot(space_df.index, space_df['Service Charge per sqft/year'],
         color=color, marker='s', linewidth=2)
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Lettable Space vs Service Charge Rate (2022-2024)',
          fontsize=14, fontweight='bold', pad=20)

# Add data labels
for year in space_df.index:
    ax1.text(year, space_df.loc[year, 'Lettable Space (sqft)'],
             f"{space_df.loc[year, 'Lettable Space (sqft)']:,} sqft",
             ha='center', va='bottom')
    ax2.text(year, space_df.loc[year, 'Service Charge per sqft/year'],
             f"KES {space_df.loc[year, 'Service Charge per sqft/year']:.2f}",
             ha='center', va='top')

fig.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Data from the Excel file
data = {
    'Year': ['2022', '2023', '2024'],
    'Lettable Space (sqft)': [95400, 101196, 101196],
    'Service Charge per sqft/year': [231.9, 260.75, 206.16]
}

# Create DataFrame and calculate metrics
df = pd.DataFrame(data)
df['Total Revenue (KES)'] = df['Lettable Space (sqft)'] * df['Service Charge per sqft/year']
df['Space Change (%)'] = df['Lettable Space (sqft)'].pct_change() * 100
df['Charge Change (%)'] = df['Service Charge per sqft/year'].pct_change() * 100
df['Revenue Change (%)'] = df['Total Revenue (KES)'].pct_change() * 100

# Formatting for display
display_df = df.copy()
display_df['Total Revenue (KES)'] = display_df['Total Revenue (KES)'].apply(lambda x: f"KES {x/1e6:.2f}M")
display_df[['Space Change (%)', 'Charge Change (%)', 'Revenue Change (%)']] = \
    display_df[['Space Change (%)', 'Charge Change (%)', 'Revenue Change (%)']].applymap(lambda x: f"{x:.1f}%" if not pd.isna(x) else "N/A")

print("Key Metrics with Percentage Changes:")
print(display_df.to_string(index=False))


# Second subplot - Revenue
plt.subplot(2, 1, 2)
bars = plt.bar(df['Year'], df['Total Revenue (KES)']/1e6, color='g')
plt.xlabel('Year')
plt.ylabel('Revenue (KES Millions)')
plt.title('Revenue Implications')

# Add revenue values and percentage changes
for i, bar in enumerate(bars):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f"KES {height:.2f}M\n({df['Revenue Change (%)'].iloc[i]:.1f}%)" if i > 0 else f"KES {height:.2f}M",
             ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
